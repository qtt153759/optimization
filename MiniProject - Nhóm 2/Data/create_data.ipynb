```python
import numpy as np
import random
import os
```

```python
def createRandomSortedList(low = 1, high = 100, size = 1):
    arr = []
    tmp = np.random.randint(low, high)
      
    for x in range(size):         
        while tmp in arr:
            tmp = random.randint(low, high)            
        arr.append(tmp)
          
    arr.sort()
      
    return np.array(arr)
```

```python
# Thu 4-7 cánh đồng 1 ngày, range = 4, không có min = 0
def create1(N, m, M, Type):
    f = open("Data/"+Type+"/"+str(N)+"-"+str(m)+"-"+str(M)+".txt", "w")
    print(N, " ", m, " ", M)
    f.write(str(N)+" "+str(m)+" "+str(M)+"\n")

    rangeDay = np.random.randint(N/7, N/4) + 1  # so ngay gioi han uoc luong
    print(rangeDay)
    # san luong thu moi ngay
    amount = np.random.randint(m, M+1, size=rangeDay)
    amount = np.sort(amount)
    print(amount)
    # so canh dong cua ngay do duoc chon random
    count = np.array([int(N/rangeDay)] * rangeDay)
    while (np.sum(count) < N):
        for i in range(rangeDay):
            # uu tien chia cac canh dong lon truoc
            day = rangeDay - 1 - i
            count[day] += 1
            if (np.sum(count) == N):
                break
    print(count)
    d = []
    s = []
    e = []
    # chia sản lượng cho các cánh đồng cùng ngày
    for day in range(rangeDay):
        mark = createRandomSortedList(1, amount[day], size=count[day]+1)
        mark[0] = 0
        mark[count[day]] = amount[day]
        for j in range(count[day]):
            d.append(mark[j+1]-mark[j])

            start = max(day-np.random.randint(4), 0)
            end = min(day+np.random.randint(4), rangeDay-1)
            # if (start<0):
            #     end-=start
            #     start=0
            # if (end>=rangeDay):
            #     start-=end-rangeDay-1
            #     end =rangeDay-1
            s.append(start)
            e.append(end)

    e = np.array(e) - min(s)
    s = np.array(s) - min(s)
    print(d)
    print(s)
    print(e)

    randomOrder = np.arange(N)
    np.random.shuffle(randomOrder)
    for i in randomOrder:
        f.write(str(d[i])+" "+str(int(s[i]))+" "+str(int(e[i]))+"\n")

    f.close()
```
