```python
import numpy as np
import time
import os
import pickle
import matplotlib.pyplot as plt
import random
import re
```

```python
TypeList = ["Type1Big","Type2Big","Type3Big","Type4Big",
             "Type1Small","Type2Small","Type3Small","Type4Small"]
```

```python
NUMBERSEED = 10 

resultTB =  np.zeros(shape=(8, 16, NUMBERSEED)) #8 Type, 16 instance
timeTB =  np.zeros(shape=(8, 16, NUMBERSEED))
resultGA =  np.zeros(shape=(8, 16, NUMBERSEED))
timeGA =  np.zeros(shape=(8, 16, NUMBERSEED))

resultIP = np.zeros(shape=(8, 16))
timeIP = np.zeros(shape=(8, 16))
resultCP = np.zeros(shape=(8, 16))
timeCP = np.zeros(shape=(8, 16))
```

```python
# Small
small_m_list = [10, 20, 35, 50]
small_M_list = [15, 30, 50, 75]
small_N_list = [10, 20, 40, 60]
# Big
big_m_list = [20, 30, 40, 50]
big_M_list = [30, 45, 60, 80]
big_N_list = [200, 400, 600, 800]
```

```python
# for type x, file i,j and method Tabu

for x in range(4):
    for i in range(4):
        # print(i)
        for j in range(4):
            if x < 4:
                instanceName = str(big_N_list[i])+"-"+str(big_m_list[j])+"-"+str(big_M_list[j])
            else:
                instanceName = str(small_N_list[i])+"-"+str(small_m_list[j])+"-"+str(small_M_list[j])
            print(instanceName)
            for seed in range(NUMBERSEED):

                fileName = "Tabu Search/Output/" + TypeList[x]+"/"+instanceName+"_SEED_"+str(seed+10)+ ".txt"
                f = open(fileName, "r")
                f.readline() # name
                time = float(f.readline().split()[1])
                result = float(f.readline().split()[1])
                resultTB[x][4*i+j][seed] = result
                timeTB[x][4*i+j][seed] = time
                f.close()

                fileName = "GA/Output/" + TypeList[x] + "/" + instanceName + "_SEED_"+str(seed)+".txt"
                f = open(fileName, "r")
                f.readline() # name
                time = float(f.readline().split()[1])
                result = float(f.readline().split()[1])
                resultGA[x][4*i+j][seed] = result
                timeGA[x][4*i+j][seed] = time
                f.close()

            fileName = "IP/Output/" + TypeList[x]+"/"+instanceName+".txt"
            f = open(fileName, "r")
            if (f.readline() == "No solution"):
                time = 600
                result = 1000
            else:
                try:
                    time = f.readline().split()[1].replace('s', '')
                    result = f.readline().split()[1]
                except:
                    time = -1
                    result = -1
            resultIP[x][4*i+j] = result
            timeIP[x][4*i+j] = time
            f.close()

            fileName= "CP/Output/"+TypeList[x]+"/"+instanceName+".txt"
            f= open(fileName,"r")
            if (f.readline() == "No solution"):
                time = 600
                result = 1000
            else:
                try:
                    time = f.readline().split()[1].replace('s', '')
                    result = f.readline().split()[1]
                except:
                    time = -1
                    result = -1
            resultCP[x][4*i+j] = result
            timeCP[x][4*i+j] = time
            f.close()
```

```python
avgResultTB = np.average(resultTB, axis=2)
avgTimeTB = np.average(timeTB, axis=2)
avgResultGA = np.average(resultGA, axis=2)
avgTimeGA = np.average(timeGA, axis=2)
```

# Đánh giá số lời giải hợp lệ


```python
count_TB = np.zeros(shape=(8))
count_GA = np.zeros(shape=(8))
count_IP = np.zeros(shape=(8))
count_CP = np.zeros(shape=(8))

for x in range(8):
    for i in range(16):
        for seed in range(NUMBERSEED):
            if (resultTB[x][i][seed] < 1000):
                count_TB[x] += 1
            if (resultGA[x][i][seed] < 1000):
                count_GA[x] += 1

        if (resultIP[x][i] < 1000):
            count_IP[x] += 1
        if (resultCP[x][i] < 1000):
            count_CP[x] += 1

print(count_IP*100/16)
print(count_CP*100/16)
print(count_TB*100/NUMBERSEED/16)
print(count_GA*100/NUMBERSEED/16)
```

```python
best = np.array(resultCP)
for x in range(8):
    for i in range(16):
        best[x][i] = min(resultCP[x][i],resultIP[x][i],np.min(resultTB[x][i]),np.min(resultGA[x][i]))
```

```python
# Số bộ đạt giá trị tối ưu
count_TB = np.zeros(shape=(8))
count_GA = np.zeros(shape=(8))
count_IP = np.zeros(shape=(8))
count_CP = np.zeros(shape=(8))
for x in range(8):
    for i in range(16):
        if (best[x][i] >= 1000):
            print(TypeList[x]+str(i))
            continue

        for seed in range(NUMBERSEED):
            if (resultTB[x][i][seed] <= best[x][i]):
                count_TB[x] += 1
            if (resultGA[x][i][seed] <= best[x][i]):
                count_GA[x] += 1
        if (resultIP[x][i] <= best[x][i]):
            count_IP[x] += 1
        if (resultCP[x][i] <= best[x][i]):
            count_CP[x] += 1

print(count_IP*100/16)
print(count_CP*100/16)
print(count_TB*100/NUMBERSEED/16)
print(count_GA*100/NUMBERSEED/16)
```

# So sánh thời gian


```python
avgTimeTabuType = np.average(avgTimeTB, axis=1)
avgTimeGAType = np.average(avgTimeGA, axis=1)
avgTimeIPType = np.average(timeIP, axis=1)
avgTimeCPType = np.average(timeCP, axis=1)
```

```python
index = [4, 5, 6, 7]

TB_time = [avgTimeTabuType[i] for i in index]
GA_time = [avgTimeGAType[i] for i in index]
IP_time = [avgTimeIPType[i] for i in index]
CP_time = [avgTimeCPType[i] for i in index]
x_name = [TypeList[i] for i in index]
```

```python
# set width of bar
barWidth = 0.8

fig = plt.subplots(figsize=(12, 8))

# Set position of bar on X axis
pos = np.arange(len(TB_time))*5

# Make the plot
plt.bar(pos-1.5, IP_time, color='tab:pink', width=barWidth,
        edgecolor='grey', label='IP')
plt.bar(pos-0.5, CP_time, color='lime', width=barWidth,
        edgecolor='grey', label='CP')
plt.bar(pos+0.5, TB_time, color='cyan', width=barWidth,
        edgecolor='grey', label='TS')
plt.bar(pos+1.5, GA_time, color='tab:orange', width=barWidth,
        edgecolor='grey', label='GA')

# Adding Xticks
plt.xlabel('Types', fontweight='bold', fontsize=15)
plt.ylabel('Running time (s)', fontweight='bold', fontsize=15)
plt.xticks(pos, x_name)
plt.grid()
plt.legend()

plt.savefig("Thực nghiệm/time_Small_data.png")
plt.show()
```
