```python
import os
import glob

output_dir = "Output"
if os.path.exists(output_dir) == False:
    os.mkdir("Output")
for i in range(1, 5):
  for path in glob.glob(f"Data/Type{i}Small*/**.txt"):
      t, name = path.split("/")[-2:]
      if os.path.exists(f"Output/{t}") == False:
          os.mkdir(f"Output/{t}")
      # print(os.getcwd())
      if os.path.exists(f"Output/{t}/{name}") == False:
          print(t, name)
          os.system(f"python CP.py --input_file Data/{t}/{name} --output_file Output/{t}/{name}")
```
