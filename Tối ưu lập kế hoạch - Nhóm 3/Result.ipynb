```python
import matplotlib.pyplot as plt
import numpy as np
import math
import json
```

```python
def read_result(path):
  with open(path) as json_file:
    data = json.load(json_file)
  return data
```

```python
def get_time_case(result_dict, case_list):
  case=[]
  time=[]
  for n, result in result_dict.items():
    if int(n) in case_list:
      case.append(n)
      time.append(math.log(result["time"]))
  return case, time
```

# Branch Bound Method

```python
branch_bound_result = read_result("/content/drive/MyDrive/Báo cáo/Tối ưu lập kế hoạch - Nhóm 3/result/branch_bound.json")
```

```python
branch_bound_result_fix={}
for n, result in branch_bound_result.items():
  new_result = {}
  new_result["time"]=np.array(result["time"]).mean()
  new_result["result"]=list(set(result["result"]))[0]
  branch_bound_result_fix[n]=new_result
```

```python
branch_bound_result_fix
```

```python
case, time = get_time_case(branch_bound_result_fix, case_list = [4,5,6,7,8,9,10])
plt.plot(case, time, marker=".", markersize=10, markerfacecolor='red')
plt.xlabel("Case")
plt.ylabel("Time (second)")
plt.show()
```

# Heuristic

## Heuristic 1: add_item == 2

```python
heuristic1_result = read_result("/content/drive/MyDrive/Báo cáo/Tối ưu lập kế hoạch - Nhóm 3/result/heuristic_add_item_2.json")
```

```python
heuristic1_result_fix={}
for n, result in heuristic1_result.items():
  new_result = {}
  new_result["time"]=np.array(result["time"]).mean()
  new_result["min_result"]=np.array(result["result"]).min()
  new_result["max_result"]=np.array(result["result"]).max()
  new_result["avg_result"]=int(np.array(result["result"]).mean())
  heuristic1_result_fix[n]=new_result
```

```python
heuristic1_result_fix
```

```python
case = []
time = []
min_result=[]
max_result=[]
avg_result=[]
for n, result in heuristic1_result_fix.items():
  case.append(int(n))
  time.append(result["time"])
  min_result.append(result["min_result"])
  max_result.append(result["max_result"])
  avg_result.append(result["avg_result"])
```

```python
case, time = get_time_case(heuristic1_result_fix, 
                           case_list = [4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 100, 200, 500, 1000, 5000, 10000, 50000, 100000, 200000])
plt.figure(figsize=(15, 5))
plt.plot(case, time, marker=".", markersize=10, markerfacecolor='red')
plt.xlabel("Case")
plt.ylabel("Time (second)")
plt.show()
```

```python
ground_truth = [1796, 5187, 6109, 4014, 3077, 5701, 9137, 9807, 15867, 21928, 35757, 66210, 141968, 304518, 718733, 3213240, 7287491, 33179184, 66355504, 138803904]
case = [4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 100, 200, 500, 1000, 5000, 10000, 50000, 100000, 200000]
min_result=[]
max_result=[]
average_result=[]
for n, result in heuristic1_result_fix.items():
  min_result.append(result['min_result'])
  max_result.append(result['max_result'])
  average_result.append(result['avg_result'])
```

```python
def compare_result(case, min, max, avg, true, begin, end):
  X_axis = np.arange(len(case[begin:end]))
  plt.bar(X_axis - 0.3, min[begin:end], 0.2, label = 'min')
  plt.bar(X_axis - 0.1, max[begin:end], 0.2, label = 'max')
  plt.bar(X_axis + 0.1, avg[begin:end], 0.2, label = 'average')
  plt.bar(X_axis + 0.3, true[begin:end], 0.2, label = 'ground truth')
  plt.xticks(X_axis, case[begin:end])
  plt.legend()
  plt.show()
```

```python
compare_result(case, min_result, max_result, average_result, ground_truth, begin=0, end=6)
```
