```python
#@markdown Code sinh data N bất kỳ
from numpy.random import randint, uniform

def generate_data(N=5, Ra=5, Rc=2, fmin=30, fmax=120, mmin=1, mmax=10):
    f = randint(fmin, fmax, size=N) # f ngẫu nhiên từ fmin đến fmax
    m = randint(mmin, mmax, size=N) # m ngẫu nhiên từ mmin đến mmax
    delta = lambda : uniform(0.85, 1.15, N) # delta R, từ -15% đến 15%
    c = (f/Rc*delta()).astype(int) # f/c có mean = Rc, variance = 15%Rc
    a = (f/Ra*delta()).astype(int) # f/a ''''''''''Ra''''''''''''''''Ra
    A = int(uniform(1.6,1.8)*sum(a*m)) # A gấp minA từ 1.6 đến 1.8 lần
    C = int(uniform(1.6,1.8)*sum(c*m)) # C ''''minC'''''''''''''''''''
    return [N, A, C, list(c), list(a), list(f), list(m)]
generate_data(5)
```

```python
import json
def write_data(n):
  data=[]
  for i in generate_data(n):
    if type(i)!=list:
      data.append(int(i))
    else:
      temp = []
      for j in i:
        temp.append(int(j))
      data.append(temp)
  with open(f'data-{n}.json', 'w') as f:
      json.dump(data, f)
```

```python
write_data(200000)
```
