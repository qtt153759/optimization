```python
def branch_bound(k):
    global N
    global A
    global C
    global c 
    global a 
    global f
    global m 
    global land 
    global cost 
    global profit 
    global best 
    global m_max 
    global x
    for i in range(m[k], m_max[k]+1):
        if land + i*a[k] <= A and cost + i*c[k] <= C:
            x[k] = i
            land += i*a[k]
            cost += i*c[k]
            profit += i*f[k]
            if k == N-1:
                if profit > best:
                    best = profit
                    # print(best, x)
            else:
                for j in range(N):
                    bo = False
                    max_withland = int((A-land)/a[j])
                    max_withcost = int((C-cost)/c[j])
                    max_rest = min(max_withland, max_withcost)
                    if profit + max_rest*f[j] > best:
                        branch_bound(k+1) 
                        bo = True
                    if bo == True:
                        break
            profit -= i*f[k]
            land -= i*a[k]
            cost -= i*c[k]
```

```python
for n in set_N:
  if n>9:
    break
  print("Case: "+str(n))
  time=[]
  result=[]
  N, A, C, c, a, f, m = read_data(n)
  m_max = []
  for i in range(N):
      A_temp = A
      C_temp = C 
      for j in range(N):
          if j != i:
              A_temp -= a[j]*m[j]
              C_temp -= c[j]*m[j]
      m_max.append(min(int(A_temp/a[i]), int(C_temp/c[i])))
  max_rate = 0
  index = 0

  for i in range(10):
    print("Runtime "+str(i))
    x = [0]*n
    land = 0
    cost = 0
    profit = 0
    best = 0
    start = timeit.default_timer()
    branch_bound(0)
    stop = timeit.default_timer()
    time.append(stop-start)
    result.append(best)
  if n not in branch_bound_result:
    branch_bound_result[n]={"time": time, "result": result}
```
