```python
from ortools.linear_solver import pywraplp
import json
import timeit
import numpy as np
```

```python
# upload file data.zip lên colab 
import shutil
set_N = [4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 100, 200, 500, 1000, 5000, 10000, 50000, 100000, 200000] # các giá trị của N, data lưu trong data-N.json
shutil.unpack_archive('data.zip', '/content', 'zip')
```

```python
def read_data(N=5): #hàm đọc data , input N, output N A C c a f m
    with open(f'data/data-{N}.json') as f:
        data = f.read()
    return json.loads(data)
```

```python
def glop(N, A, C, c, a, f, m):
    solver = pywraplp.Solver.CreateSolver('GLOP')
    infinity = solver.infinity()

    k = {}
    for i in range(N):
        k[i] = solver.IntVar(m[i], infinity, f'k_{i}')

    solver.Add(sum((k[i]*c[i]) for i in range(N)) <= C)
    solver.Add(sum((k[i]*a[i]) for i in range(N)) <= A)

    solver.Maximize(sum((k[i]*f[i]) for i in range(N)))
    status = solver.Solve()

    if status == pywraplp.Solver.OPTIMAL:
        # print('Runtime: '+str(stop - start))
        # print('Solution:')
        # print('Objective value =', solver.Objective().Value())
        result=[]
        for i in range(N):
            result.append(int(k[i].solution_value()))

        return sum(np.array(f)*np.array(result))

    else:
        print('The problem does not have an optimal solution.')
```

```python
glop_result = {}

for n in set_N:
  if n>200000:
    break
  print("Case :"+str(n))
  N, A, C, c, a, f, m = read_data(n)
  time_list=[]
  result_list=[]
  for i in range(10):
    print("Runtime: "+str(i))
    start = timeit.default_timer()
    result = glop(N, A, C, c, a, f, m)
    stop = timeit.default_timer()
    time_list.append(stop-start)
    result_list.append(result)
  if n not in glop_result:
    glop_result[n]={"time": time_list, "result": result_list}
```
