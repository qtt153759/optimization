```python
import numpy as np
import numpy.random as npr
import json
import timeit
```

```python
# upload file data.zip lên colab 
import shutil
set_N = [4, 5, 6, 7, 8, 9, 10, 15, 20, 30, 50, 100, 200, 500, 1000, 5000, 10000, 50000, 100000, 200000] # các giá trị của N, data lưu trong data-N.json
shutil.unpack_archive('data.zip', '/content', 'zip')
```

```python
def read_data(N=5): #hàm đọc data , input N, output N A C c a f m
    with open(f'data/data-{N}.json') as f:
        data = f.read()
    return json.loads(data)
```

- add_item: số loại sản phẩm sẽ được thêm
- num_item_add: số lượng sản phẩm thêm sau mỗi vòng lặp
- change_num_item: số vòng lặp mà ra kết quả giống nhau để giảm num_item_add xuống
- early_break: số vòng lặp mà ra kết quả giống nhau để có thể dừng sớm

```python
def check_result(C, A, c, a, q):
    if sum(np.array(c)*np.array(q))<C and sum(np.array(a)*np.array(q))<A:
        return True
    else:
        return False
def main(N, A, C, c, a, f, m, trial = 1000000, add_item=2, num_item_add=1000, change_num_item=10, early_break=10):
    q = m.copy()
    temp = np.array(f)-(np.array(a)+np.array(c))
    # choosen_index = np.argsort(temp)[int(-N/2-1):]
    choosen_index = np.argsort(temp)[-add_item:]
    roulette_wheel = 1/temp[choosen_index]
    roulette_wheel = roulette_wheel/sum(roulette_wheel)
    # print(roulette_wheel)
    for i in range(trial):
        # print(q)
        # index = npr.choice(int(N/2+1), p=roulette_wheel)
        index = npr.choice(add_item, p=roulette_wheel)
        temp = q.copy()
        temp[choosen_index[index]]+=num_item_add
        if check_result(C, A, c, a, temp):
          count=0
          q = temp
        else:
          if 'count' not in locals():
            count=0
            continue
          count+=1
        if count==change_num_item and num_item_add!=1:
          num_item_add=int(num_item_add/2)
          count=0
        if count==early_break and num_item_add==1:
          break
    # print("Objective value: "+str(sum(np.array(f)*np.array(q))))
    # print(q)
    return sum(np.array(f)*np.array(q))
```

```python
heuristic_result = {}

for n in set_N:
  if n>200000:
    break
  print("Case :"+str(n))
  N, A, C, c, a, f, m = read_data(n)
  time_list=[]
  result_list=[]
  for i in range(10):
    print("Runtime: "+str(i))
    start = timeit.default_timer()
    result = main(N, A, C, c, a, f, m)
    stop = timeit.default_timer()
    time_list.append(stop-start)
    result_list.append(result)
  if n not in heuristic_result:
    heuristic_result[n]={"time": time_list, "result": result_list}
```

```python
def change(heuristic_result):
  new_heuristic_result = {}
  for n, result in heuristic_result.items():
    new_result={}
    new_result["time"]=result["time"]
    new_result["result"]=[int(i) for i in result["result"]]
    new_heuristic_result[int(n)] = new_result
  return new_heuristic_result
```
