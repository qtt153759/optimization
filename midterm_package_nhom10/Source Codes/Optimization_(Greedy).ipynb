```python
import sys
import time

f = open('/content/data_5_3.txt', 'r')
[N, M] = [int(x) for x in f.readline().split()]

Q = []
for i in range(0, N):
    Q_ = [int(x) for x in f.readline().split()]
    Q.append(Q_)

d = []
for i in range(0, M+1):
    d_ = [int(x) for x in f.readline().split()]
    d.append(d_)
    
q = [int(x) for x in f.readline().split()]


#số hàng cần lấy phải nhỏ hơn số hàng hiện có trong kho
for i in range(0,N):
    if(q[i] <= sum(Q[i])):
        continue
    else:
        print('STOP\nCheck input value!!')
        sys.exit()
```

# **1. THAM LAM 1: ĐẾN NƠI TỔNG HÀNG THIẾU NHIỀU NHẤT**
Idea: 
1. Tính sumq(i): tổng hàng max có thể lấy khi đến kệ i. Với điều kiện: 
+ Kệ nào đã đi qua thì sumq(i) = 0
+ Tại kệ i, nếu số hàng có trên kệ ít hơn số hàng cần lấy thì ta lấy hết, và ngược lại, nếu số hàng hiện có nhiều hơn số hàng cần lấy thì ta chỉ lấy đúng số lượng cần lấy thôi 
2. Tìm index phần tử max của sumq(i), khi đó k tiếp theo = index tìm được, sau đó update q[i]
3. Điều kiện dừng: khi lấy hết hàng (q.count(0) == N). Sau đó tính tổng quãng đường đã đi


```python
#biến
sumq = [0]*(M)        #mảng lưu tổng hàng trên từng kệ
q_ = [0]*(N)             
k_ = []               #mảng lưu lộ trình
k_.append(0)

print('q({0}) = {1}'.format(N,q), end='-->')
#tính tổng hàng có thể  lấy trên mỗi kệ
def sum_q(q_,i,k):
    for j in range(0,N):
        if(Q[j][i] < q[j]):
            q_[j] = Q[j][i]
        else:
            q_[j] = q[j]
    return sum(q_)

#tìm index tiếp theo
def q_max(Q,q,k):
    for i in range(0,M):
        sumq[i] = sum_q(q_,i,k)

    for i in range(0,len(k_)):
        if(k_[i] != 0):
            sumq[k_[i]-1] = 0
    
    k = sumq.index(max(sumq)) + 1
    return k

def TRY(k):
    #bắt đầu tại điểm 0
    k = q_max(Q,q,k)
    k_.append(k)
    update_q(k)
    TRY(k)
  
#ktra số hàng còn lại sau mỗi lẫn di chuyển
def update_q(k):
  for i in range (0,N):
      if (q[i] >= Q[i][k-1]):
          q[i] = q[i] - Q[i][k-1]
      else:
          q[i] = 0
  print(q, end='-->')

  if(q.count(0) == N):
      #về cửa kho
      k_.append(0)
      s = 0
      for i in range(0, len(k_)-1):
          s = s + d[k_[i]][k_[i+1]]
      
      print('STOP\nk = ', k_, sep='')
      print('s_min =', s)
      end_time = time.time()
      print('Time:', (end_time - start_time))
      sys.exit()

start_time = time.time()
TRY(0)
```

# **2. THAM LAM 2: ĐẾN NƠI GẦN ĐIỂM XUẤT PHÁT NHẤT**
Idea: 
1. x[ ]: là mảng lưu hàng i của ma trận khoảng cách d[M+1][M+1], trong đó:
+ khoảng cách từ kệ k đến chính nó = 99999
+ khoảng cách kệ k đến những kệ đã đi qua = 99999
2. Tìm index phần tử min của x[], khi đó k tiếp theo = index tìm được, sau đó update q[i]
3. Điều kiện dừng: khi lấy hết hàng (q.count(0) == N). Sau đó tính tổng quãng đường đã đi

```python
#biến
v = [False]*(M+1)     #lưu vết
x = [99999]*(M+1)     #mảng lưu hàng i ma trận d[M+1][M+1]
k_ = []               #mảng lưu lộ trình
k_.append(0)

print('q({0}) = {1}'.format(N,q), end='-->')

#tìm index tiếp theo
def s_min(x,k):
    i = x.index(min(x))
    
    if(v[i]==False):
        k = i
    else:
        x[x.index(min(x))] = 99999
        k = x.index(min(x))
    
    return k         

def TRY(k):
    #bắt đầu tại điểm 0
    v[k] = True
    for i in range(0,M+1):
        x[i]=d[k][i]
        x[k]=99999
        for i in range (0,len(k_)): #chỉ có 1 chiều đi ra từ kệ i
          x[k_[i]] = 99999

    
    k = s_min(x,k)
    v[k]= True
    k_.append(k)
    update(k)
    TRY(k)
  
#ktra số hàng còn lại sau mỗi lẫn di chuyển
def update(k):
  for i in range (0,N):
      if (q[i] >= Q[i][k-1]):
          q[i] = q[i] - Q[i][k-1]
      else:
          q[i] = 0
  print(q, end='-->')
  if(q.count(0) == N):
      #về cửa kho
      k_.append(0)
      s = 0
      for i in range(0, len(k_)-1):
          s = s + d[k_[i]][k_[i+1]]
      
      print('STOP\n', k_, sep='')
      print('s_min =', s)
      end_time = time.time()
      print('Time:', (end_time - start_time))
      sys.exit()

start_time = time.time()
TRY(0)
```

# **3.THAM LAM 3: ĐẾN NƠI TỔNG HÀNG CÒN THIẾU/KC MAX**
Idea: (tổng hợp của 2 hướng trước: tổng hàng max / khoảng cách min)
1. Tính sumq(i): tổng hàng max/khoảng cách min. Với điều kiện:
+ Kệ nào đã đi qua thì sumq(i) = 0
+ Tại kệ i, nếu số hàng có trên kệ ít hơn số hàng cần lấy thì ta lấy hết, và ngược lại, nếu số hàng hiện có nhiều hơn số hàng cần lấy thì ta chỉ lấy đúng số lượng cần lấy thôi
+ khoảng cách trong sumq(i) là d[k][i]
2. Tìm index phần tử max của sumq(i), khi đó k tiếp theo = index tìm được, sau đó update q[i]
3. Điều kiện dừng: khi lấy hết hàng (q.count(0) == N). Sau đó tính tổng quãng đường đã đi
